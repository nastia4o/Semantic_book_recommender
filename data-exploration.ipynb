```python
import numpy as np
# LOOK AT THE BOOKS WITH MISSING DESCRIPTION  (CORE)
#books that are missing description put one in that column NOT missing 0
books["missing_description"]= np.where(books["description"].isna(), 1,0)
# age of
books["age_of book"]= 2024 - books["published_year"]
```

```python
#list with all the columns that we are interested in
columns_of_interest = ["num_pages", "age_of book","missing_description","average_rating"]
#we create a  correlation matrix
#method spearman -> more appropriate when you're dealing with non continuous
correlation_matrix = books[columns_of_interest].corr(method="spearman")
#convert into a seaborn heatmap
sns.set_theme(style="white")
plt.figure(figsize=(8,6))
##to explain???
heatmap = sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman Correlation"})
#set the tittle
heatmap.set_title("Correlation Heatmap")
plt.show()
#everything close to 1 or -1 is a really strong correlation
#anything closer to 0 is a weak correlation  (there's no really relationship)
```

```python
# delete those observation with missing values
#let's find how many books
books[(books["description"].isna())|
      (books["published_year"].isna())|
      (books["num_pages"].isna())|
      (books["published_year"].isna())
]
#it's less than 5 percent we can get rid of this information
```

```python
#~ -> negation
book_missing = books[~(books["description"].isna())&
      ~(books["published_year"].isna())&
      ~(books["num_pages"].isna())&
      ~(books["published_year"].isna())
]
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
#not natural breakpoint
#we need to create an arbitrary cut point ourselves
# how many  words a description needs to have to be meaningful
#pandas code filter pandas dataframe books with description 1 and 4 words
book_missing.loc[book_missing["words_in_description"].between(1,4),"description"]
# we can notice that is too small and skip that (1,4)
```

```python
#create some filtering and eliminate all the words that
# we create a new dataframe and filter
book_missing_25_words = book_missing[book_missing["words_in_description"]>=25]
book_missing_25_words
#now we have a base for our recommender
```

```python
#create new field and pair the two if exists
book_missing_25_words["title_and_subtitle"]= (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             # we convert it into string there's a lot of missing values and they are considered as a  float in pandas
             book_missing_25_words[['title','subtitle']].astype(str).agg(": ".join,axis=1))

)
book_missing_25_words
```

```python
# create a new description column that tags a description with unique identifier
# we need to do that because
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13","description"]].astype(str).agg(" ".join,axis=1)
book_missing_25_words
```

```python
#dataset is now clean we can save it to file
#drop the fields that we don't need
(
    book_missing_25_words
    .drop(['subtitle','missing_description','age_of book','words_in_description'],axis=1)
    #save into csv and  don't save index
    .to_csv("books_cleaned.csv",index=False)
)
```
